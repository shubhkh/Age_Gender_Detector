#### Blurring

```python
img1=cv2.imread("Resources/Data/damon.png.jpg")
img=cv2.cvtColor(img1,cv2.COLOR_BGR2HSV)
figure_size=7
result=cv2.blur(img,(figure_size,figure_size))
plt.figure(figsize=(11,6))

plt.subplot(121),plt.imshow(cv2.cvtColor(img,cv2.COLOR_HSV2BGR)),plt.title("Original")
plt.xticks([]),plt.yticks([])

plt.subplot(122),plt.imshow(cv2.cvtColor(result,cv2.COLOR_HSV2BGR)),plt.title("Blurred")
plt.xticks([]),plt.yticks([])
plt.show()
```

#### Zoom

```python
img1=cv2.imread("Resources/Data/lena.png.jpg")
```

```python
(height,width)=img1.shape[0:2]
zoom=cv2.resize(img1,None,fx=1.5,fy=1.5,interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom)
```

```python
zoom=cv2.resize(img1,(int(width*2),int(height*2)),interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom)
```

# Rotate

```python
m=cv2.getRotationMatrix2D((width/2,height/2),30,1)
rotate=cv2.warpAffine(img1,m,(height,width))
plt.imshow(rotate)
```
